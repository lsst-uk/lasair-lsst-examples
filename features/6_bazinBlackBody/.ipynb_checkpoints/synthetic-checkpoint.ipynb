```python
import importlib  
BBBcode = "lasair-lsst.pipeline.filter.features.BBBEngine"
BBBEngine = importlib.import_module(BBBcode)
```

```python
mjdStart = 60000
A = 1000
T = 5             # 5000 kelvin
kr = 0.3          # magnitudes per day
t0 = 4
kf = 0.1
npoint = 10      # number of detections
cadence = 2      # days between detections
```

```python
bands = ['u', 'g', 'r', 'i', 'z', 'y']
WL    = [0.380,     0.500,     0.620,     0.740,     0.880,     1.000, ]

def blackbody(wl, T):
    hck = 14.387
    return np.power(wl, -3.0) / (np.exp(hck/(wl*T)) - 1)

def bazin(t):
    tau = t - t0
    ef = np.exp(-kf*tau)
    er = np.exp(-kr*tau)
    return ef/(1+er)

def expit(t):
    return np.exp(kr*t)
```

```python
dsList = []
for i in range(npoint):
    t = float(i)       # detection every day
    whichBand = random.randrange(6)
    band    = bands[whichBand]
    wavelength = WL[whichBand]
#    flux = A * blackbody(wavelength, T) * expit(t)
    flux = A * blackbody(wavelength, T) * bazin(t)
    dsList.append({
        'band':band, 
        'midpointMjdTai':mjdStart + cadence*t, 
        'psfFlux':flux,
        'psfFluxErr': 1,
    })

#for ds in dsList:
#    print(ds)

alert = {
    'diaObject':{'diaObjectId':12324}, 
    'diaSourcesList': dsList, 
    'diaForcedSourcesList':[]
}
```

```python
BE = BBBEngine.BBB('LSST', nforced=4, ebv=0, A=100, T=4, t0=0, kr=0.1, kf=0.01, verbose=False)
fit =  BE.make_fit(alert)
print(fit)
```

```python
if fit:
    filename = '../image/%s_synthetic.png' % str(alert['diaObject']['diaObjectId'])
    BE.plot(alert, fit, filename, size=5)
    display(Image(filename=filename))
```
