## 2D Lightcurve Fitting: Introduction

### `Exponential*Blackbody` fit and `Bazin*Blackbody` fit

This notebook shows how the BazinBlackBody features work. It attempts two 2D fits: one models
the flux as exponential in time and blackbody in  wavelength, the other as bazin in time 
(exp rise and exp fall) and blackbody in  wavelength. One (with exponential) is an 
explosion in the early stages, the other (with Bazin) is an explosion that is fading from its peak.

Please read the introduction below, then try some of the notebooks that utilise the fitting library:

rise rate and fall rate of 0.13 and  0.01 magnitudes per day respectively.

converges even with no range in the wavelength direction.

### Fitting a 2D surface 
Here we plot the two models that we attempt to fit to the multiband lightcurve: 
the `Exponential*Blackbody` and the `Bazin*Blackbody`.
The axes are wavelength (microns) and time (days). 
The functions are exponential in time (linear in magnitude) times blackbody; 
and Bazin in time (exp rise then exp fall) times blackbody.

The following functions define the 2D models that we are fitting, 
the exponential-blackbody and the bazin-blackbody.

```python
import json, sys, math
import numpy as np
from IPython.display import Image, display
from mpl_toolkits.mplot3d import Axes3D  
import matplotlib.pyplot as plt

T = 5   # Just choose arbitrary temperature 5,000 Kelvin
def blackbody(wl, T):
    hck = 14.387
    q = np.exp(hck/(wl*T))
    return np.power(wl, -3.0) /(q - 1)
def bazinBB(tau, wl):
    T = 10
    kr = 1
    kf = 0.3
    ef = np.exp(-kf*tau)
    er = np.exp(-kr*tau)
    return blackbody(wl, T) * ef/(1+er)
def expBB(tau, wl):
    T = 10
    k = 0.7
    er = np.exp(k*tau)
    return blackbody(wl, T) * er
```

3D plotting from `mplot3d`

```python
fig = plt.figure(figsize=(12, 8))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')
ax1.azim = ax2.azim = 40
tau = np.arange(-3, 3, 0.5)
wl  = np.arange(0.1, 2.0, 0.05)
Tau, Wl = np.meshgrid(tau, wl)

zs = np.array(expBB(np.ravel(Tau), np.ravel(Wl)))
Z = zs.reshape(Tau.shape)
ax1.plot_surface(Tau, Wl, Z, edgecolors='k', alpha=.6)
ax1.set_xlabel('Time')
ax1.set_ylabel('Wavelength')
ax1.set_title('Exponential*Blackbody')
ax1.view_init(elev=15., azim=140)

zs = np.array(bazinBB(np.ravel(Tau), np.ravel(Wl)))
Z = zs.reshape(Tau.shape)
ax2.plot_surface(Tau, Wl, Z, edgecolors='k', alpha=.6)
ax2.set_xlabel('Time (days)')
ax2.set_ylabel('Wavelength (micron)')
ax2.set_title('Bazin*Blackbody')
ax2.view_init(elev=15., azim=140)

plt.show()
```

## Extinction Correction
Note that the code used by Lasair makes an **extinction correction** to the fluxes before
trying to fit the model. A source reddened by galactic dust is not really as red as we see it.

## Model Selection
The `Exponential*Blackbody` model represents a source of constant temperature whose 
magnitude (log of flux) is rising linearly. This model has three parameters:
- `A`: overall flux scale
- `T`: temperature in kilo-Kelvins (T=4 means 4,000 K)
- `kr`: the rise rate per day, so that flux ~ `exp(kf*t)`

Note that a rise rate of 1 per day is equivalent to 1.086 magnitudes per day, 
so to within a few percent we can think of `kr` as rise rate in magnitude per day.

The `Bazin*Blackbody` model represents 
a linear rise in magnitude followed by a linear fall in magnitude.
The latter model has five parameters: 
- `A`: overall flux scale
- `T`: temperature in kilo-Kelvins (T=4 means 4,000 K)
- `t0`: time of peak brightness relative to the time of the first lightcurve point
- `kr`: the rise rate per day before the peak
- `kf`: the fall rate per day after the peak

If the fitting converges for both models, and the quality of fit (sum of squares of residuals) 
is the same for both, then we select the simpler model (`Exponential*Blackbody`). Otherwise,
we use the [Akaike information criterion](https://en.wikipedia.org/wiki/Akaike_information_criterion)
to select the *most efficient* fit, considering residuals and number of parameters.
