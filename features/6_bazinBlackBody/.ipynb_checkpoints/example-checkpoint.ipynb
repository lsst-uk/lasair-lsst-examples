This notebook shows how the BazinBlackBody features work. It attempts two 2D fits: one models
the flux as exponential in time and blackbody in  wavelength, the other as bazin in time 
(exp rise and exp fall) and blackbody in  wavelength. 
Note that exponential rise of flux is linear rise of magnitude.

```python
import json, sys, math
from IPython.display import Image, display
```

```python
import importlib  
BBBcode = "lasair-lsst.pipeline.filter.features.BBBEngine"
BBBEngine = importlib.import_module(BBBcode)
```

```python
# Here is a fake alert
file = 'SNIbc_312228'
alert = json.loads(open('../json/%s.json'%file).read())
```

The arguments to the function are:
- Whether it is a ZTF or LSST system
- the number of forced detections to include, if available
- the extinction E(B-V) to use before making the fit
- the initial parameters as defined in [the introduciton](https://github.com/lsst-uk/lasair-examples/blob/main/notebooks/features/6_bazinBlackBody/introduction.ipynb).
- verbossity of output
If the fit converges, the return is either the fit from the 3-parameter 
exponential-blackbody, or from the 5-parameter bazin-blackbody. 
The fit dictionary carries the fitting parameters.

```python
BE = BBBEngine.BBB('LSST', nforced=4, ebv=0, A=100, T=4, t0=6, kr=0.1, kf=0.01, verbose=False)
fit =  BE.make_fit(alert)
print(fit)
```

```python
if fit:
    filename = '../image/%s_example.png'%file
    BE.plot(alert, fit, filename, size=5)
    display(Image(filename=filename))
```
